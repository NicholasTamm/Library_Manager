```python
%%sql
CREATE TRIGGER prevent_employee_deletion
BEFORE DELETE ON Employee
FOR EACH ROW
WHEN EXISTS (
    SELECT 1 FROM Event WHERE hostID = OLD.employeeID
)
BEGIN
    SELECT RAISE(ABORT, 'Cannot delete an employee who is hosting an event');
END;
```
